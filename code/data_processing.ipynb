```python
import pandas as pd
import numpy as np
```

```python
bkash = pd.read_csv(r'D:\FinTech Bangladesh\data\bkash_most_relevant_2020_2026.csv')
```

```python
bkash.shape
```

```python
bkash = bkash.iloc[:500]
```

```python
bkash.shape
```

```python
from transformers import pipeline

pipe = pipeline("sentiment-analysis", 
                model="nlptown/bert-base-multilingual-uncased-sentiment",
                device=-1,  # use 0 for GPU
                truncation=True)

# Batch processing — much faster than row-by-row
results = pipe(bkash['content'].tolist(), batch_size=32)
bkash['text_sentiment'] = [r['label'] for r in results]
```

```python
bkash['rating_sentiment'] = bkash['score'].apply(
    lambda s: 'Positive' if s>=4 else ('Neutral' if s==3 else 'Negative')
)

# Find disagreements
disagreements = bkash[bkash['text_sentiment'] != bkash['rating_sentiment']]
print(f"Disagreement rate: {len(disagreements)/len(bkash)*100:.1f}%")
print(disagreements[['content','score','rating_sentiment','text_sentiment']].head(20))
```
