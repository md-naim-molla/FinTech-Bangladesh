```python
import pandas as pd
import time
from tqdm import tqdm
from google import genai

# ========= CONFIG =========
API_KEY = "AIzaSyDahc37Oq7hqLAGAE6SCckbR5sfsw6_TvA"
INPUT_FILE = r"D:\FinTech Bangladesh\code\nagad_50.csv"
OUTPUT_FILE = "reviews_with_sentiment.csv"
TEXT_COLUMN = "content"
MODEL = "gemini-2.0-flash-lite"

BATCH_SIZE = 5
RETRY_LIMIT = 3
SLEEP_BETWEEN_CALLS = 1
# ==========================


client = genai.Client(api_key=API_KEY)


def clean_label(label):
    label = label.strip().lower()
    if label in ["positive", "neutral", "negative"]:
        return label
    return "unknown"


def analyze_batch(texts):
    """
    Sends multiple reviews in one prompt.
    Returns list of sentiments.
    """

    joined_reviews = "\n".join(
        [f"{i+1}. {t}" for i, t in enumerate(texts)]
    )

    prompt = f"""
Classify each review as positive, neutral, or negative.

Return ONLY labels separated by commas.
Do not explain.

Reviews:
{joined_reviews}
"""

    for attempt in range(RETRY_LIMIT):
        try:
            response = client.models.generate_content(
                model=MODEL,
                contents=prompt
            )

            raw = response.text.strip()

            results = [clean_label(x) for x in raw.split(",")]

            # If mismatch, retry
            if len(results) != len(texts):
                raise ValueError("Mismatch count")

            return results

        except Exception as e:
            if attempt == RETRY_LIMIT - 1:
                print("Failed batch:", e)
                return ["error"] * len(texts)

            time.sleep(2)


# ===== LOAD DATA =====
df = pd.read_csv(INPUT_FILE)

reviews = df[TEXT_COLUMN].astype(str).tolist()

all_sentiments = []

# ===== PROCESS BATCHES =====
for i in tqdm(range(0, len(reviews), BATCH_SIZE)):
    batch = reviews[i:i+BATCH_SIZE]

    sentiments = analyze_batch(batch)
    all_sentiments.extend(sentiments)

    time.sleep(SLEEP_BETWEEN_CALLS)


# ===== SAVE RESULT =====
df["sentiment"] = all_sentiments
df.to_csv(OUTPUT_FILE, index=False)

print("Done. Saved →", OUTPUT_FILE)
```
